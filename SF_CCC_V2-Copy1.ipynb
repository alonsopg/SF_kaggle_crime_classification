2. Use FP growth to mine association rules

```python
import Orange
data = Orange.data.Table("/Users/user/Jupyter/datasets/SFCCC_dataset/categorical_vals.csv")
```

```python
data.domain.attributes
```

```python
data.domain.class_var
#DiscreteVariable('Category', values=['Descript', 'DayofWeek', 'PdDistrict', 'Resolution', 'Address'])
```

```python
import Orange
print("Attributes:", ", ".join(x.name for x in data.domain.attributes))
print("Class:", data.domain.class_var.name)
print("Data instances", len(data))

target = "soft"
print("Data instances with %s prescriptions:" % target)
atts = data.domain.attributes
for d in data:
    if d.get_class() == target:
        print(" ".join(["%14s" % str(d[a]) for a in atts]))
```

```python
from orangecontrib.associate.fpgrowth import *  
X, mapping = OneHot.encode(data, include_class=True)
```

```python
X
```

```python
sorted(mapping.items())
```

```python
itemsets = dict(frequent_itemsets(X, .4))
```

```python
len(itemsets)
```

```python
 class_items = {item
                for item, var, _ in OneHot.decode(mapping, data, mapping)
                if var is data.domain.class_var}
```

```python
sorted(class_items)
```

```python
data.domain.class_var.values
```

```python
rules = [(P, Q, supp, conf)
         for P, Q, supp, conf in association_rules(itemsets, .8)
         if len(Q) == 1 and Q & class_items]
len(rules)
```

```python
rules
```

```python
names = {item: '{}={}'.format(var.name, val)
         for item, var, val in OneHot.decode(mapping, data, mapping)}
for ante, cons, supp, conf in rules[:5]:
    print(', '.join(names[i] for i in ante), '-->',
          names[next(iter(cons))],
          '(supp: {}, conf: {})'.format(supp, conf))
```
